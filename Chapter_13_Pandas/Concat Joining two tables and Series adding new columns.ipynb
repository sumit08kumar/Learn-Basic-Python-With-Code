```python
import pandas as pd

india_weather = pd.DataFrame({
    "City": ["Mumbai", "Delhi","Banglore", "Bihar"],
    "Temperature": [32, 25, 20, 29],
    "Humidity": [80, 60, 78, 50]
})
india_weather
```

```python
usa_weather = pd.DataFrame({
    "City": ["New York", "Los Anglis","Chicago", "Orlando"],
    "Temperature": [21, 14, 35, 19],
    "Humidity": [68, 65, 75, 51]
})
usa_weather
```

```python
df =pd.concat([india_weather, usa_weather])
df
```

```python
df =pd.concat([india_weather, usa_weather], ignore_index=True )
df
```

```python
df =pd.concat([india_weather, usa_weather], keys=["india", "usa"] )
df
```

```python
temperature_df = pd.DataFrame({
    "City": ["Mumbai", "Delhi", "Banglore", "Bihar"],
    "Temperature": [32, 45 ,25, 20]
})
temperature_df
```

```python
windspeed_df = pd.DataFrame({
    "City": ["Mumbai", "Delhi", "Banglore", "Bihar"],
    "Windspeed": [7, 12, 9, 10]
})
windspeed_df
```

```python
df = pd.concat([temperature_df, windspeed_df], axis=1)
df
```

```python
temperature_df = pd.DataFrame({
    "City": ["Mumbai", "Delhi", "Banglore", "Bihar"],
    "Temperature": [32, 45 ,25, 20]
}, index=[0,1,2,3])
temperature_df
```

```python
windspeed_df = pd.DataFrame({
    "City": ["Delhi", "Mumbai", "Banglore"],
    "Windspeed": [7, 12, 9]
}, index=[1,0,2])
windspeed_df
```

```python
df = pd.concat([temperature_df, windspeed_df], axis=1)
df
```
