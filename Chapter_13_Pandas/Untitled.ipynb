```python
def convert_people_cell(cell):
    if cell == "n.a.":
        return 'Sumit Kumar Singh'
    return cell

df = pd.read_csv("D:\Excel\Pandas Datasets\Stocks1.csv", converters = {
    'people' : convert_people_cell
})
df
```

```python
def convert_eps_cell(cell):
    if cell == "Not Available":
        return None
    return cell

df = pd.read_csv("D:\Excel\Pandas Datasets\Stocks1.csv", converters = {
    'eps' : convert_eps_cell
})
df
```

```python
df.to_excel("Stocks_data2.xlsx",sheet_name="Stocks")
```

```python
df_stocks = pd.DataFrame({
    'tickers': ['Google', 'IBM', 'Microsoft'],
    'price': [845,65,64],
    'pe': [30.37, 14.26, 30.12],
    'eps': [27.852, 4.61, 2.122]
})

df_weather = pd.DataFrame({
    'day': ['1/1/2022', '2/1/2022', '3/1/2022'],
    'temperature': [32,25,20],
    'event': ['Rainy', 'Sunny', 'Cloudy']
})
```

```python
with pd.ExcelWriter('stocks_weather.xlsx') as writer:
    df_stocks.to_excel(writer, sheet_name="stocks")
    df_weather.to_excel(writer, sheet_name="weather")
```
