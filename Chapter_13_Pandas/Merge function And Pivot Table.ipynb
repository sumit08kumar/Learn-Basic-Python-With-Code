```python
import pandas as pd
df2 = pd.DataFrame({
    "City":["Chicago", "New York", "Orlando", "Gujarat"],
    "Humidity": [65,68,75, 70],
})
df2
```

```python
df1 = pd.DataFrame({
    "City":["New York", "Chicago", "Orlando", "Mumbai"],
    "Humidity": [21,14,32, 25],
})
df1
```

```python
df3 = pd.merge(df1,df2, on="City")
df3
```

```python
df3 = pd.merge(df1,df2, on="City", how="outer")
df3
```

```python
df3 = pd.merge(df1,df2, on="City", how = "inner")
df3
```

```python
df3 = pd.merge(df1,df2, on="City", how="left")
df3
```
